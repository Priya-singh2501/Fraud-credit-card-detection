UNDER SAMPLING

```python
#number of fraud transaction ---> 232
```

```python
legit_sample = legit.sample(n=232)
```

```python
dv = pd.concat([legit_sample, fraud] , axis=0)
```

```python
dv.head(10)
```

```python
dv.tail(8)
```

```python
dv['Class'].value_counts()
```

```python
dv.groupby('Class').mean()
```

SPLITTING THE DATA INTO FEATURE AND TARGET

```python
X = dv.drop(columns='Class', axis=1)
Y = dv['Class']
```

```python
print(X)
```

```python
print(Y)
```

SPLIT THE DATA INTO  TRAINING DATA NAD TESTING DATA

```python
X_train, X_test, Y_train, Y_test = train_test_split(X ,Y , test_size =0.2 ,stratify=Y , random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

MODEL TRAINING

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train , X_train_prediction)
print('Accuracy score of training data :',training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test , X_test_prediction)
print('Accuracy score of test data :',test_data_accuracy)
```
